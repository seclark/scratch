#L_AD = 80 ions:#
##highly anisotropic with strong magnetic field along the x axis##

```python
ppvna004_2alf6_coldens = fits.getdata("ppvna004_.2alf6/ppvna004_.2alf6_coldens.fits")
ppvna004_2alf6_coldens_rht0 = fits.getdata("ppvna004_.2alf6/ppvna004_.2alf6_coldens_xyt00.fits")
ppvna004_2alf6_coldens_rht1 = fits.getdata("ppvna004_.2alf6/ppvna004_.2alf6_coldens_xyt01.fits")
ppvna004_2alf6_coldens_rht2 = fits.getdata("ppvna004_.2alf6/ppvna004_.2alf6_coldens_xyt02.fits")
fig = plt.figure(figsize = (12, 12))
ax1 = fig.add_subplot(221)
ax1.set_title("total column")
ax1.imshow(ppvna004_2alf6_coldens, cmap="Greys")
ax2 = fig.add_subplot(222)
ax2.imshow(ppvna004_2alf6_coldens_rht0, cmap="Greys")
ax2.set_title("w 55, s 15")
ax3 = fig.add_subplot(224)
ax3.imshow(ppvna004_2alf6_coldens_rht1, cmap="Greys")
ax3.set_title("w 35, s 2")
ax4 = fig.add_subplot(223)
ax4.imshow(ppvna004_2alf6_coldens_rht2, cmap="Greys")
ax4.set_title("w 45, s 2")
#fits.getheader("ppvna004_.2/ppvna004_.2_coldens_xyt02.fits")
```

##Some examples, with wlen = 45 and smr = 2:##

```python
lad80_1_ions_data = fits.getdata("ppvna004_.2alf6/ppvna004_.2alf6_2.fits")
lad80_1_ions = fits.getdata("ppvna004_.2alf6/ppvna004_.2alf6_2_xyt01.fits")
lad80_1_neutrals_data = fits.getdata("ppvne004_.2alf6/ppvne004_.2alf6_2.fits")
lad80_1_neutrals = fits.getdata("ppvne004_.2alf6/ppvne004_.2alf6_2_xyt01.fits")

lad40_1_ions_data = fits.getdata("ppvna004_.2/ppvna004_.2_3.fits")
lad40_1_ions = fits.getdata("ppvna004_.2/ppvna004_.2_3_xyt01.fits")
lad40_1_neutrals_data = fits.getdata("ppvne004_.2/ppvne004_.2_3.fits")
lad40_1_neutrals = fits.getdata("ppvne004_.2/ppvne004_.2_3_xyt01.fits")

b1p05_1_ions_data = fits.getdata("b.1p.05_v=.2_ppv/b.1p.05_v=.2_ppv_0.fits")
b1p05_1_ions = fits.getdata("b.1p.05_v=.2_ppv/b.1p.05_v=.2_ppv_0_xyt01.fits")

fig = plt.figure(figsize = (15, 15))

ax0 = fig.add_subplot(3, 4, 1)
ax0.imshow(lad80_1_ions_data, cmap="Greys")
ax0.set_title(r"$L_{AD}$ 80, vel 2, ions")
ax1 = fig.add_subplot(3, 4, 2)
ax1.imshow(lad80_1_ions, cmap="Greys")
ax1.set_title(r"$L_{AD}$ 80, vel 2, ions")
ax2 = fig.add_subplot(3, 4, 3)
ax2.imshow(lad80_1_neutrals_data, cmap="Greys")
ax2.set_title(r"$L_{AD}$ 80, vel 2, neutrals")
ax3 = fig.add_subplot(3, 4, 4)
ax3.imshow(lad80_1_neutrals, cmap="Greys")
ax3.set_title(r"$L_{AD}$ 80, vel 2, neutrals")

ax4 = fig.add_subplot(3, 4, 5)
ax4.imshow(lad40_1_ions_data, cmap="Greys")
ax4.set_title(r"$L_{AD}$ 40, vel 3, ions")
ax5 = fig.add_subplot(3, 4, 6)
ax5.imshow(lad40_1_ions, cmap="Greys")
ax5.set_title(r"$L_{AD}$ 40, vel 3, ions")
ax6 = fig.add_subplot(3, 4, 7)
ax6.imshow(lad40_1_neutrals_data, cmap="Greys")
ax6.set_title(r"$L_{AD}$ 40, vel 3, neutrals")
ax7 = fig.add_subplot(3, 4, 8)
ax7.imshow(lad40_1_neutrals, cmap="Greys")
ax7.set_title(r"$L_{AD}$ 40, vel 3, neutrals")

ax8 = fig.add_subplot(3, 4, 9)
ax8.imshow(b1p05_1_ions_data, cmap="Greys")
ax8.set_title(r"single fluid, vel 1, data")
ax9 = fig.add_subplot(3, 4, 10)
ax9.imshow(b1p05_1_ions, cmap="Greys")
ax9.set_title(r"single fluid, vel 1, RHT")
```
