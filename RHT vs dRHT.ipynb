# RHT and dRHT with Juan's simulated clouds

```python
%pylab inline --no-import-all
from IPython.display import Image
np.random.seed(0)
np.set_printoptions(precision=2)
```

```python
from astropy.io import fits
```

Ran the RHT and dRHT with the same parameters on Juan's cloud. The following were run with -wlen=51 -smr=5 -frac=0.7

```python
orig = fits.getdata("new_fits.fits")
data4 = fits.getdata("new_fits_xyt04.fits")
data5 = fits.getdata("new_fits_xyt05.fits")
```

```python
fig = plt.figure(figsize=(18, 4))
ax1 = fig.add_subplot(131)
cbar = ax1.imshow(np.log10(orig), cmap="Greys")
fig.colorbar(cbar)
ax1.set_xlim(200,800)
ax1.set_ylim(50, 650)
ax1.set_title("log(simulation data)")

ax2 = fig.add_subplot(132)
ax2.imshow(data5, cmap="Greys")
ax2.set_xlim(200,800)
ax2.set_ylim(50, 650)
ax2.set_title("original RHT")

ax3 = fig.add_subplot(133)
ax3.imshow(data4, cmap="Greys")
ax3.set_xlim(200,800)
ax3.set_ylim(50, 650)
ax3.set_title("directional RHT")
```

For two runs with the same window length, the directional RHT is definitely better at picking up curves, since you only have to get over the threshold within a window *radius* rather than a window diameter in order to be recorded.

As designed, both the RHT and the dRHT are detecting linear structure without regard for the density regime. I think it's worth asking what we want to extract from the data at this point.
